# Simple Graph - LangGraph Basics

This notebook demonstrates the basics of creating a simple graph using LangGraph.

```python
# Import required libraries
from langgraph.graph import Graph, StateGraph
from langchain_core.messages import HumanMessage, AIMessage
from typing import TypedDict, Annotated
import operator
```

```python
# Define your graph state
class GraphState(TypedDict):
    messages: Annotated[list, operator.add]
    
```

```python
# Define nodes
def node_1(state: GraphState):
    print("---Node 1---")
    return {"messages": ["Node 1 executed"]}

def node_2(state: GraphState):
    print("---Node 2---")
    return {"messages": ["Node 2 executed"]}
```

```python
# Build the graph
workflow = StateGraph(GraphState)

# Add nodes
workflow.add_node("node_1", node_1)
workflow.add_node("node_2", node_2)

# Add edges
workflow.set_entry_point("node_1")
workflow.add_edge("node_1", "node_2")
workflow.set_finish_point("node_2")

# Compile the graph
app = workflow.compile()
```

```python
# Run the graph
result = app.invoke({"messages": []})
print("\nFinal Result:")
print(result)
```
